 # Delayed Flights Logistic Regression

```python
# Initial imports
import matplotlib.pyplot as plt
import pandas as pd
```

# First Run

```python
# load data flight from amazon aws 
data = "https://airport-efficiency.s3.amazonaws.com/Merged_cleaned_airport_data.csv"
df = pd.read_csv(data)
df.head()
```

```python
#list the types to find out which ones need to be changed using get_dummies 
df.dtypes
```

```python
#need to resorve the difference of catergoies: ua_code share and ua should be in the same column 
df.BRANDED_CODE_SHARE.value_counts()
```

```python
# merge "codeshares" to their corresponding airlines 
df.loc[(df.BRANDED_CODE_SHARE == "UA_CODESHARE"), "BRANDED_CODE_SHARE"] = "UA"
df.loc[(df.BRANDED_CODE_SHARE == "AA_CODESHARE"), "BRANDED_CODE_SHARE"] = "AA"
df.loc[(df.BRANDED_CODE_SHARE == "DL_CODESHARE"), "BRANDED_CODE_SHARE"] = "DL"
df.loc[(df.BRANDED_CODE_SHARE == "AS_CODESHARE"), "BRANDED_CODE_SHARE"] = "AS"
```

```python
#make sure our there are no more "airline codeshares" in our dataset 
df.BRANDED_CODE_SHARE.value_counts()
```

```python
#make sure our 3 airlines are in our dataset 
df.ORIGIN.value_counts()
```

```python
# count all the states in our data 
df.DEST_STATE_NM.value_counts()
```

```python
#divide the states into five regions 
northeast = ['Massachusetts',
'Rhode Island',
'Connecticut',
'Vermont',
'New Hampshire',
'Maine',
'Pennsylvania',
'New Jersey',
'New York'
]

souteast = ['Georgia',
'North Carolina',
'South Carolina',
'Virginia',
'West Virginia',
'Kentucky',
'Tennessee',
'Mississippi',
'Alabama',
'Delaware',
'Maryland',
'Florida',
'Louisiana',
'Arkansas'
]

midwest = ['Minnesota',
'Wisconsin',
'Illinois',
'Ohio',
'Indiana',
'Michigan',
'Missouri',
'Iowa',
'Kansas',
'Nebraska',
'North Dakota',
'South Dakota'
]

southwest = ['New Mexico',
'Arizona',
'Oklahoma',
'Texas'
]

west = ['California',
'Colorado',
'Nevada',
'Hawaii',
'Alaska',
'Oregon',
'Utah',
'Idaho',
'Montana',
'Wyoming',
'Washington'
]

noncontinental = ['U.S. Virgin Islands',
'Puerto Rico'
]
```

```python
# this code created a new column "region" and classifies each state to its corresponding region 
regions=pd.DataFrame([northeast,souteast,midwest,southwest,west,noncontinental],index=['northeast','souteast','midwest','southwest','west','noncontinental'])
regions=regions.reset_index().melt('index')
df['Region'] = df['DEST_STATE_NM'].map(dict(zip(regions['value'],regions['index'])))
```

```python
#double check the code above classified the states into their corresponding regions 
df.Region.value_counts()
```

```python
#make sure that there are no null values
df.isna().sum()
```

```python
# drop more columns to make sure it does not overlap with delays 
drop = ['DEP_TIME','DEP_DELAY_GROUP', 'DEST_STATE_NM', 'DEST' ]
```

```python
#create a new df without the dropped columns 
df_dropped = df.drop( drop , 1)
```

```python
df_dropped.dtypes
```

```python
# use get_dummies for the brand column 
brand_df = pd.get_dummies(df_dropped, columns=["BRANDED_CODE_SHARE"])
brand_df
```

```python
# use get_dummies for the origin column 
origin_df = pd.get_dummies(brand_df, columns=["ORIGIN"])
origin_df
```

```python
# use get_dummies for the region column 
final_df = pd.get_dummies(origin_df, columns=["Region"])
final_df
```

```python
# make sure our get dummies worked 
final_df.dtypes
```

```python
# change date to date time 
final_df['DATE'] = pd.to_datetime(final_df['DATE'], format='%Y-%m-%d')
```

```python
 
final_df['DAY_OF_MONTH'] = final_df.DATE.dt.day
```

```python
#final drop = date 
final_df = final_df.drop('DATE', 1)
```

```python
# get dummies for departure time 
final_df = pd.get_dummies(final_df, columns=["DEP_TIME_BLK"])
```

 # separate the features (x) from the target (y) 

```python
# define features set 
y = final_df["DELAYED"]
X = final_df.drop(columns=["DELAYED"])
```

 # Split our data into training and testing

```python
# Splitting into Train and Test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
#coun thte y test value 
y_test.value_counts()
```

# Feature Scaling

```python
# scale data 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

 # Create a Logistic Regression Model

```python
# create a logistic regression model 
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1, max_iter=1000)
classifier
```

 # Fit (train) or model using the training data

```python
classifier.fit(X_train, y_train)
```

 # Make predictions

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

# Validate the model using the test data

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score, accuracy_score
bacc = (balanced_accuracy_score(y_test, predictions))
acc = accuracy_score(y_test,predictions)
print('Balanced Accuracy Score: ' + str(round(bacc, 4)))
print('Accuracy Score: ' + str(round(acc, 4)))
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predictions, labels=[1, 0]))
```

```python
# Display results 
cmtx = pd.DataFrame(
    confusion_matrix(y_test, predictions, labels=[1, 0]), 
    index=['Actual Delayed', 'Actual On-Time'], 
    columns=['Predicted Delayed', 'Predicted On-Time']
    )
cmtx
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, predictions))
```

```python
importance = classifier.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d , Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```
